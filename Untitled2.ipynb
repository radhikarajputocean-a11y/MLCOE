```python
#top 10 crops avg yield
```

```python
avg_yield_by_crop = df.groupby("Crop")["Yield (kg/ha)"].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
avg_yield_by_crop.plot(kind="bar", color="skyblue")
plt.title("Top 10 Crops by Average Yield (kg/ha)")
plt.ylabel("Yield (kg/ha)")
plt.xticks(rotation=45)
plt.show()
```

```python
#top 10  states by avg production
```

```python
avg_prod_by_state = df.groupby("State")["Production (tonnes)"].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
avg_prod_by_state.plot(kind="bar", color="lightgreen")
plt.title("Top 10 States by Average Production (tonnes)")
plt.ylabel("Production (tonnes)")
plt.xticks(rotation=45)
plt.show()
```
